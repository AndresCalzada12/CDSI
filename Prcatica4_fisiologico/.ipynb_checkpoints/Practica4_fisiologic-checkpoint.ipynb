```python
import heartpy as hp
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import welch
```

Es importante mencionar que las señales que fueron graficadas son señales HRV (Variabilidad de la Frecuencia Cardíaca), es evidente que las señales obtenidas son señales que presentn ruido, por lo que una vez que visualicemos la señales el sigueinte paso es de hacer un filtrado de ellas. Debido a que para esta practica vamos a utilizar las señales lo más impias posibles. 

```python
#definimos una  frecuencia de muestreo, la cual sera la misma para todas las señales
sample_rate = 250
#cargamos el archivo .txt para concer la morfología de la señal por medio de la libreria de numpy.
# Como la libreria de heartpy recibe datos de tipo numpy no abra que hacer un procesamiento extra para poder ejecutar el código.
data_1 = np.loadtxt('Datos/ANDRES.txt', delimiter=',')
data_2 = np.loadtxt('Datos/ARIANA.txt', delimiter=',')
data_3 = np.loadtxt('Datos/DAFNE.txt', delimiter=',')
data_4 = np.loadtxt('Datos/JESUS.txt', delimiter=',')
data_5 = np.loadtxt('Datos/RACHEL.txt', delimiter=',')
data_6 = np.loadtxt('Datos/SCARLETT.txt', delimiter=',')

# Calcular el tiempo correspondiente a cada muestra si se conoce la frecuencia de muestreo
tiempo1 = np.arange(len(data_1)) / sample_rate
tiempo2 = np.arange(len(data_2)) / sample_rate
tiempo3 = np.arange(len(data_3)) / sample_rate
tiempo4 = np.arange(len(data_4)) / sample_rate
tiempo5 = np.arange(len(data_5)) / sample_rate
tiempo6 = np.arange(len(data_6)) / sample_rate

#Crear un subplot de 4 columnas y 2 filas 
fig, axs = plt.subplots(2, 3, figsize=(20,10))

# Asignar datos a cada subplot
axs[0, 0].plot(tiempo1, data_1, 'tab:blue')
axs[0, 0].set_title('HRV Andres')

axs[0, 1].plot(tiempo2, data_2, 'tab:orange')
axs[0, 1].set_title('HRV Ariana')

axs[0, 2].plot(tiempo3, data_3, 'tab:green')
axs[0, 2].set_title('HRV Dafne')

axs[1, 0].plot(tiempo4, data_4, 'tab:red')
axs[1, 0].set_title('HRV Jesus')

axs[1, 1].plot(tiempo5, data_5, 'tab:purple')
axs[1, 1].set_title('HRV Rachel')

axs[1, 2].plot(tiempo6, data_6, 'tab:brown')
axs[1, 2].set_title('HRV Scarlett')

# Ajustar espaciado
plt.tight_layout()

# Mostrar el plot
plt.show()
```

#### *<span style = "color : #01A9DB" >Espectro en la frecuencia de las señales de HRV</span>*

```python
#Señal 1
# Calcular la FFT
fft_result = np.fft.fft(data_1)
# Calcular las frecuencias correspondientes
frequencies = np.fft.fftfreq(len(fft_result), d=1/sample_rate)
# Tomar solo las frecuencias positivas (la mitad debido a la simetría)
positive_frequencies_1 = frequencies[:len(frequencies)//2]
magnitude_spectrum_1 = np.abs(fft_result[:len(fft_result)//2])

#Señal 2
# Calcular la FFT
fft_result = np.fft.fft(data_2)
# Calcular las frecuencias correspondientes
frequencies = np.fft.fftfreq(len(fft_result), d=1/sample_rate)
# Tomar solo las frecuencias positivas (la mitad debido a la simetría)
positive_frequencies_2 = frequencies[:len(frequencies)//2]
magnitude_spectrum_2 = np.abs(fft_result[:len(fft_result)//2])

#Señal 3
# Calcular la FFT
fft_result = np.fft.fft(data_3)
# Calcular las frecuencias correspondientes
frequencies = np.fft.fftfreq(len(fft_result), d=1/sample_rate)
# Tomar solo las frecuencias positivas (la mitad debido a la simetría)
positive_frequencies_3 = frequencies[:len(frequencies)//2]
magnitude_spectrum_3 = np.abs(fft_result[:len(fft_result)//2])

#Señal 4
# Calcular la FFT
fft_result = np.fft.fft(data_4)
# Calcular las frecuencias correspondientes
frequencies = np.fft.fftfreq(len(fft_result), d=1/sample_rate)
# Tomar solo las frecuencias positivas (la mitad debido a la simetría)
positive_frequencies_4 = frequencies[:len(frequencies)//2]
magnitude_spectrum_4 = np.abs(fft_result[:len(fft_result)//2])

#Señal 5
# Calcular la FFT
fft_result = np.fft.fft(data_5)
# Calcular las frecuencias correspondientes
frequencies = np.fft.fftfreq(len(fft_result), d=1/sample_rate)
# Tomar solo las frecuencias positivas (la mitad debido a la simetría)
positive_frequencies_5 = frequencies[:len(frequencies)//2]
magnitude_spectrum_5 = np.abs(fft_result[:len(fft_result)//2])

#Señal 6
# Calcular la FFT
fft_result = np.fft.fft(data_6)
# Calcular las frecuencias correspondientes
frequencies = np.fft.fftfreq(len(fft_result), d=1/sample_rate)
# Tomar solo las frecuencias positivas (la mitad debido a la simetría)
positive_frequencies_6 = frequencies[:len(frequencies)//2]
magnitude_spectrum_6 = np.abs(fft_result[:len(fft_result)//2])
```

```python
# Crear un subplot de 2 filas y 3 columnas
fig, axs = plt.subplots(2, 3, figsize=(20, 10))

# Lista de tus datos de frecuencias y magnitudes
positive_frequencies_list = [positive_frequencies_1, positive_frequencies_2, positive_frequencies_3, positive_frequencies_4, positive_frequencies_5, positive_frequencies_6]
magnitude_spectrum_list = [magnitude_spectrum_1, magnitude_spectrum_2, magnitude_spectrum_3, magnitude_spectrum_4, magnitude_spectrum_5, magnitude_spectrum_6]
titles = ['FFT HRV Andres', 'FFT HRV Ariana', 'FFT HRV Dafne', 'FFT HRV Jesus', 'FFT HRV Rachel', 'FFT HRV Scarlett']

# Iterar sobre cada subplot para asignar los datos
for i, ax in enumerate(axs.flat):
    # Encuentra el índice donde la frecuencia es igual o justo mayor a 20 Hz
    idx_20Hz = np.argmax(positive_frequencies_list[i] >= 20)
    
    # Dibuja solo hasta ese índice para obtener hasta 20 Hz
    ax.plot(positive_frequencies_list[i][:idx_20Hz], magnitude_spectrum_list[i][:idx_20Hz])
    ax.set_title(titles[i])
    ax.set_xlim([0, 20])  # Asegura que el eje x muestre hasta 20 Hz
    ax.set_xlabel('Frecuencia (Hz)')
    ax.set_ylabel('Magnitud')

# Ajustar espaciado
plt.tight_layout()

# Mostrar el plot
plt.show()
```

#### *<span style = "color : #01A9DB" >Filtrado de las señales</span>*

Dado que la información relevante en una señal de HRV se encuentra en los rangos de frecuencia siguientes:

* VLF (Very Low Frequency): Esta banda de frecuencia va de 0.0033 a 0.04 Hz. Se asocia con influencias térmicas y hormonales en el ritmo cardíaco, aunque su significado exacto y mecanismos subyacentes aún están bajo investigación.

* LF (Low Frequency): La banda LF abarca de 0.04 a 0.15 Hz. Se considera que refleja tanto la actividad simpática como la parasimpática, aunque tradicionalmente se ha asociado más con la actividad simpática. La potencia en esta banda a menudo se interpreta en relación con la modulación de la presión arterial y la función barorrefleja.

* HF (High Frequency): Esta banda va de 0.15 a 0.4 Hz y está asociada principalmente con la modulación parasimpática del corazón. La HF refleja la variabilidad del ritmo cardíaco relacionada con el ciclo respiratorio (respiración sinusal).

Dado que primero vamos a eliminar la linea base de nuestra señal HRV vamos a utilizar un filtro tipo notch con una frecuencia de corte de 0.05


```python
filtered_1 = hp.filter_signal(data_1, cutoff = 0.05, sample_rate = sample_rate, filtertype='notch')
filtered_2 = hp.filter_signal(data_2, cutoff = 0.05, sample_rate = sample_rate, filtertype='notch')
filtered_3 = hp.filter_signal(data_3, cutoff = 0.05, sample_rate = sample_rate, filtertype='notch')
filtered_4 = hp.filter_signal(data_4, cutoff = 0.05, sample_rate = sample_rate, filtertype='notch')
filtered_5 = hp.filter_signal(data_5, cutoff = 0.05, sample_rate = sample_rate, filtertype='notch')
filtered_6 = hp.filter_signal(data_6, cutoff = 0.05, sample_rate = sample_rate, filtertype='notch')

# Calcular los nuevos tiempos a cada muestra si se conoce la frecuencia de muestreo
tiempo1 = np.arange(len(filtered_1)) / sample_rate
tiempo2 = np.arange(len(filtered_2)) / sample_rate
tiempo3 = np.arange(len(filtered_3)) / sample_rate
tiempo4 = np.arange(len(filtered_4)) / sample_rate
tiempo5 = np.arange(len(filtered_5)) / sample_rate
tiempo6 = np.arange(len(filtered_6)) / sample_rate
```

```python
#Crear un subplot de 3 columnas y 2 filas 
fig, axs = plt.subplots(2, 3, figsize=(20,10))

# Asignar datos a cada subplot
axs[0, 0].plot(tiempo1[0:200], filtered_1[0:200], 'tab:blue')
axs[0, 0].set_title('HRV Andres (filtrada)')

axs[0, 1].plot(tiempo2[0:200], filtered_2[0:200], 'tab:orange')
axs[0, 1].set_title('HRV Ariana (filtrada)')

axs[0, 2].plot(tiempo3[0:200], filtered_3[0:200], 'tab:green')
axs[0, 2].set_title('HRV Dafne (filtrada)')

axs[1, 0].plot(tiempo4[0:200], filtered_4[0:200], 'tab:red')
axs[1, 0].set_title('HRV Jesus (filtrada)')

axs[1, 1].plot(tiempo5[0:200], filtered_5[0:200], 'tab:purple')
axs[1, 1].set_title('HRV Rachel (filtrada)')

axs[1, 2].plot(tiempo6[0:200], filtered_6[0:200], 'tab:brown')
axs[1, 2].set_title('HRV Scarlett (filtrada)')

# Ajustar espaciado
plt.tight_layout()

# Mostrar el plot
plt.show()
```

# *<span style = "color : #01A9DB" >Análisis de Variabilidad de la Frecuencia Cardíaca</span>*

Una vez que se filtraron las señales por medio del filtro "notch", para solamente liminar la desviación base en las señales de ECG. Vamos a comenzar con el análisis de las señales filtradas. Para este análisis vamos a realizar lo siguiente.

Donde vamos a realizar el estudio de las variaciones de los intervalos de tiempo entre los latidos del corazón, conocidos como intervalos RR. Dado que el HRV es un indicador poderoso de la activdad del sistema nervioso central (SNA) y su balance entre los componentes simpaticos y parasimpático, que regula las respuestas del cuerpo al estres y al reposo.

El análisis de HRV ha ganado importancia debido a su capacidad para proporcionar insights sobre la salud cardiovascular, el estrés, la recuperación, y el bienestar general. Es particularmente relevante en contextos como el de tu proyecto, donde se examinan las respuestas fisiológicas a situaciones potencialmente estresantes, como las exposiciones públicas.

#### *<span style = "color : #01A9DB" >Aplicaciones del Análisis de HRV</span>*

En esta práctica vamos a realizar lo siguiente con las señales de HRV obtenidas de los participantes en los momentos cuando se realizó la presentación de los proyectos para el curso de CDSI.

Identificación de Respuesta al Estrés: Comprarando las métricas de HRV durante las exposciones con periodos de reposo, se pueden indentificar los cambios fisiológicos para tratar de indentificar estres.

Correlacion con Eventos Durante la Exposición: Analizando como cambian las métricas de HRV en momentos especificos (por ejemplo, al comenzar a hablar, durante preguntas)

Para análizar estos cambios bruscos y tendencias en los cuales podemos bservar si exsite algun cambio brusco o algún evento vamos a realizar lo siguiente.

* Detectar Cambios Bruscos: Picos

Vamos a utilizar la funcion find_peaks de scipy.signal. Esta función puede ser ajustada para detectar picos por encima de cierto umbral y con ciertas propiedades (como la prominencia), lo que nos ayuda a identificar picos significativos que podrían indicar eventos de estrés o nerviosismo.

* Identificar Tendencias

Para identificar tendencias, vamos a aplicar técnicas de suavizado y luego buscar segmentos de la señal que muestren una pendiente positiva o negativa sostenida.

```python
import numpy as np
from scipy.signal import find_peaks
from scipy.ndimage import gaussian_filter1d
```

### **<span style="color:orange">Análisis para Andres</span>**

```python
# Detectar picos altos
peaks, _ = find_peaks(filtered_1, height=None) 

# Tendencias: podemos usar un suavizado y luego calcular la derivada
smoothed_signal = gaussian_filter1d(filtered_1, sigma=2)
```

```python
# Acomodando las dos gráficas en un subplot de 2 columnas y una fila
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Gráfico de señal HRV con picos y valles
axs[0].plot(filtered_1, label='HRV Señal')
axs[0].scatter(peaks, filtered_1[peaks], color='red', label='Picos')
axs[0].set_title('Señal HRV con Picos')
axs[0].set_xlabel('Tiempo')
axs[0].set_ylabel('HRV')
axs[0].legend()

# Gráfico de tendencias de señal HRV
axs[1].plot(filtered_1, label='Señal Original de HRV', alpha=0.5)
axs[1].plot(smoothed_signal, label='Señal HRV Suavizada', color='cyan')
axs[1].set_title('Tendencias de la Señal HRV')
axs[1].set_xlabel('Tiempo')
axs[1].set_ylabel('HRV')
axs[1].legend()

plt.tight_layout()
plt.show()
```

Ahora vamos a trabajar con la señal suavizada, vamos a realizar un pequeño ajuste para que podamos manejar un umbral de donde queremos observar donde existe cambio en la señal, es decir, conociendo que la frecuencia normal de una persona es de 60-100 lpm, una persona que se encuentra estresada o nerviosa eleva su ritmo cardiaco, por lo que en este caso umbrales superiores a los 100 lpm van a ser considerados evento.

Convertir la señal HRV a una frecuencia cardíaca estimada: Primero, necesitamos convertir la señal de HRV (que probablemente esté en milisegundos entre latidos, o en alguna otra medida directamente relacionada con la variabilidad de la frecuencia cardíaca) a una frecuencia cardíaca estimada en latidos por minuto. Esto se puede hacer con la fórmula básica para convertir intervalos RR (en milisegundos) a FC en lpm: 

## $FC= \frac{60000}{RR}$

Para este análisis vamos a trabajar con las señales que ya estan suavizadas, ya que en estas es más fácil de observar los picos y los cambios.

```python
#Estimación de la frecuencia cardiaca a latidos por minuto

fc_estimada = 60000 / smoothed_signal  

#Identificar picos y luego filtrar por aquellos que representan un cambio significativo
peaks, properties = find_peaks(fc_estimada, height=140)  
plt.figure(figsize=(14, 6))
plt.plot(fc_estimada, label='Estimación del Ritmo Cardiaco')
plt.scatter(peaks, fc_estimada[peaks], color='red', label='Cambios Significativos de HR (>140 BPM)')
plt.title('Señal HRV con Cambios Significativos en HR')
plt.xlabel('Tiempo')
plt.ylabel('Estimación del Ritmo Cardiaco (BPM)')
plt.legend()
plt.show()
```

# **<span style="color:#0000FF">2. Análisis en el dominio de la $Frecuencia$</span>**

Una vez que ya se realizo un análisis en el dominio del tiempo, existen otras alternativas para poder análizar las señales, una de ellas es análizar las señales en el dominio de la frecuencia. Para poder realizar un análisis espectral de potencia de una señal HRV, vamos a utilizar la transformada rápida de Fourier (FFT) o métodos específicos de análisis espectral como la densidad espectral de potencia (PSD) para descomponer la señal en sus componentes de frecuencia.

Esto permite examinar la distribución de la energía de la señal a lo largo de diferentes frecuencias y nos puede ser útil para análizar señales de HRV, ya que en las diferentes bandas de frecuencia podemos asociar distintos aspectos de la regulación autonómica del corazón, es decir, podemos observar cambios cuando la persona tuvo un evento.


```python
# Calcular las diferencias entre intervalos RR consecutivos (delta RR)
delta_rr_intervals = np.diff(filtered_1)

# Utilizar Welch's method para estimar la densidad espectral de potencia (PSD)
frequencies, psd = welch(delta_rr_intervals, fs=4.0, nperseg=256, noverlap=128)

# Visualización del PSD para identificar las bandas de frecuencia de HRV
plt.figure(figsize=(10, 6))
plt.semilogy(frequencies, psd)
plt.title('HRV Power Spectral Density (PSD)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Potencia Espectral')
plt.xlim(0, 0.5) 
plt.ylim(min(psd), max(psd))
plt.show()
```
